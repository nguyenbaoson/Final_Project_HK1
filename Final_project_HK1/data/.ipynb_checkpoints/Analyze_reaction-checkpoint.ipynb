# Ti·ªÅn x·ª≠ l√Ω d·ªØ li·ªáu:

T√°ch c·ªôt reactions v√† g·ªôp v·ªõi post_text, shares,comments, reactions_count

```python
raw_df['reactions'] = raw_df['reactions'].apply(lambda x: dict(ast.literal_eval(x)) if isinstance(x, str) else {})
raw_df_reactions = raw_df['reactions'].apply(pd.Series )
```

```python
raw_df_with_reactions = pd.concat([raw_df, raw_df_reactions], axis=1).drop('reactions', axis=1)
print(raw_df_with_reactions[['post_text','th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                                 'shares','comments','reaction_count']])
```

Thay th·∫ø gi√° tr·ªã NaN trong c√°c c·ªôt b·∫±ng gi√° tr·ªã 0

```python
columns_to_replace = ['post_text', 'th√≠ch', 'y√™u th√≠ch', 'haha', 'wow', 'bu·ªìn', 'ph·∫´n n·ªô', 'th∆∞∆°ng th∆∞∆°ng', 'shares', 'comments', 'reaction_count']
```

```python
for column in columns_to_replace:
    raw_df_with_reactions[column].fillna(0, inplace=True)

result_df = raw_df_with_reactions[['post_text', 'th√≠ch', 'y√™u th√≠ch', 'haha', 'wow', 'bu·ªìn', 'ph·∫´n n·ªô', 'th∆∞∆°ng th∆∞∆°ng', 'shares', 'comments', 'reaction_count']]
result_df
```

T√≠nh t·ªïng c·ªßa t·ª´ng c·ªôt

```python
sum_columns = result_df[['th√≠ch', 'y√™u th√≠ch', 'haha', 'wow', 'bu·ªìn', 'ph·∫´n n·ªô', 'th∆∞∆°ng th∆∞∆°ng', 'shares', 'comments', 'reaction_count']].sum()
sum_columns
```

V·∫Ω bi·ªÉu ƒë·ªì th·ªÉ hi·ªán gi√° tr·ªã c·ªßa t·ª´ng c·ªôt gi√° tr·ªã

```python
plt.figure(figsize=(12, 6))
sum_columns.plot(kind='bar', color='skyblue')
plt.xlabel('Y·∫øu t·ªë')
plt.ylabel('T·ªïng gi√° tr·ªã')
plt.title('Bi·ªÉu ƒë·ªì c·ªôt cho gi√° tr·ªã sum_columns')
plt.show()
```

Hi·ªÉn th·ªã m·ªëi quan h·ªá th√≠ch,y√™u th√≠ch,haha,wow,bu·ªìn,ph·∫´n n·ªô,th∆∞∆°ng th∆∞∆°ng v·ªõi c·ªôt Shares

```python
raw_df_with_reactions[['th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng']].corrwith(raw_df_with_reactions['shares'])
```

V·∫Ω bi·ªÉu ƒë·ªì ƒë∆∞·ªùng c·ªßa t·ª´ng lo·∫°i reactions

```python
plt.figure(figsize=(12, 6))
colors = sns.color_palette("husl", 7)
for i, column in enumerate(['th√≠ch', 'y√™u th√≠ch', 'haha', 'wow', 'bu·ªìn', 'ph·∫´n n·ªô', 'th∆∞∆°ng th∆∞∆°ng']):
    plt.plot(raw_df_with_reactions[column], label=column, marker='o', color=colors[i])
plt.gca().xaxis.set_major_formatter(DateFormatter('%d-%m-%y %H:%M'))
plt.xlabel('Th·ªùi gian')
plt.ylabel('S·ªë l∆∞·ª£ng')
plt.title('Bi·ªÉu ƒë·ªì ƒë∆∞·ªùng theo th·ªùi gian c·ªßa c√°c c·ªôt')
plt.legend()
plt.show()
```
