```python
import ast
```

```python
df['reactions'] = df['reactions'].apply(lambda x: dict(ast.literal_eval(x)) if isinstance(x, str) else {})
raw_df_reactions = df['reactions'].apply(pd.Series )
```

```python
df_with_reactions = pd.concat([df, raw_df_reactions], axis=1).drop('reactions', axis=1)
print(df_with_reactions[['post_text','th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                                 'shares','comments','reaction_count']])
```
