# Tiền Xử Lý Dữ Liệu:

Tách cột reactions và gộp với post_text, shares,comments, reactions_count

```python
import ast
```

```python
df['reactions'] = df['reactions'].apply(lambda x: dict(ast.literal_eval(x)) if isinstance(x, str) else {})
raw_df_reactions = df['reactions'].apply(pd.Series )
```

```python
df_with_reactions = pd.concat([df, raw_df_reactions], axis=1).drop('reactions', axis=1)
print(df_with_reactions[['post_text','thích','yêu thích','haha','wow','buồn','phẫn nộ','thương thương',
                                 'shares','comments','reaction_count']])
```

Thay thế giá trị NaN trong các cột bằng giá trị 0

```python
columns_to_replace = ['post_text', 'thích', 'yêu thích', 'haha', 'wow', 'buồn', 'phẫn nộ', 'thương thương', 'shares', 'comments', 'reaction_count']
```

```python
for column in columns_to_replace:
    df_with_reactions[column].fillna(0, inplace=True)

result_df = df_with_reactions[['post_text', 'thích', 'yêu thích', 'haha', 'wow', 'buồn', 'phẫn nộ', 'thương thương', 'shares', 'comments', 'reaction_count']]
result_df
```
